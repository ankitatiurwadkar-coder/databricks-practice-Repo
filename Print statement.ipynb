## Creating dataframe from Cloud Storage


##Using access KEys

```python
from pyspark.sql.types import *
CUSTOM_SCHEMA = StructType([
StructField("id", IntegerType()),
StructField("age", IntegerType()),
StructField("gen", StringType()),
StructField("desig", StringType()),
StructField("salary", IntegerType()),

])

spark.read.csv(path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/sample_dataset/users_001.csv",
header=True,
schema=CUSTOM_SCHEMA).display()
```

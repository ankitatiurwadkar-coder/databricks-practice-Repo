##Prerequisites

```python
df = spark.read.csv(
    "/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/employee.csv",
    header=True,
    inferSchema=True,
    sep="|",
    quote="'"
)
df.display()
```

##Problem Statement
>Fetch only the names of the employees whose gen='M'

```python
df.select(col("name")).filter(col("gen")=='M').display()
```

##Lower

```python
from pyspark.sql.functions import lower
# df.filter(lower(col("company"))=="cisco").display()
df.filter(lower("company")=="cisco").display()
```

##GROUPING,AGGRE, SORT

```python
df.groupBy("gen").count().display()
```

##Derive New columns

```python
from pyspark.sql.functions import lit
df.withColumn('is_Alive',lit('True')).display()
```

```python
from pyspark.sql.functions import when
df.withColumn(
    "category",
    when(col("exp")>=10, "Senior")
    .when(col("exp")>=5, "Mid level")
    .when(col("exp")>=0, "Junior")
    .otherwise("Interen"),
).limit(4).display()
```

```python
df.selectExpr(
"*",
"""
CASE
    WHEN exp>=10 THEN 'Senior'
    WHEN exp>=5 THEN 'Mid Level'
    WHEN exp>=0 THEN 'Junior'
    ELSE 'Invalid Exper'
END AS category
"""

).display()
```

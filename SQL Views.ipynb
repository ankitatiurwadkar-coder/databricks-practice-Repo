```python
df = spark.read.csv(
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/employee.csv",
    header=True,
    inferSchema=True,
    quote="'",
    sep="|"
)

df.display()
```

Views 
>>In Spark, Views are temporary virtual tables that can be created from Dataframes

##Temporary Views

##Create View

```python
df.createOrReplaceTempView("emp_vw")
```

##Accessing View

```python
%sql
select id, name from emp_vw
```

##Display all temporary Views

```python
for view in spark.catalog.listTables():
  print(view)
```

##Global View


```python
df.createOrReplaceGlobalTempView("emp_glb_vw")
```

```python
%sql
select name,id from global_temp.emp_glb_vw;
```

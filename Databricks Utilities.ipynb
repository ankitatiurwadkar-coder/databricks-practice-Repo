```python
dbutils.widgets.combobox(label='top_n_records',
                         name='param_combo_top_n_records',
                         choices=['1','2','3','4','5'],
                         defaultValue='1'

)

```

```python
dbutils.widgets.dropdown(label='top_n_records_field',
                         name='param_combo_top_n_records_field',
                         choices=['production','Supply_chain','mechanical'],
                         defaultValue='production'

)
```

```python
dbutils.widgets.get('param_combo_top_n_records_field')
```
